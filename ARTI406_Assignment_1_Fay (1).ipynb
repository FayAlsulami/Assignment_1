EDA is the first and most important step in any Machine Learning project.
Before building models, we must understand:

- What does the data represent?
- Are there missing values?
- Are there outliers?
- What patterns exist?
- Which variables influence others?

If we do not understand the data, we cannot build a good model.

### Why EDA is Important

In real-world machine learning projects:

    1- 70–80% of the time is spent on understanding and cleaning data
    2- Only 20–30% is spent building models

**EDA helps us:**

- Detect errors
- Identify trends
- Discover relationships
- Make business decisions

##DATE CLEANING

```python
print("PERFORMING DATA CLEANING")

rows_before = df.shape[0]

# Remove duplicates
df_clean = df.drop_duplicates()

# Remove rows with missing values
df_clean = df_clean.dropna()

rows_after = df_clean.shape[0]

print(f"\nRows before cleaning: {rows_before}")
print(f"Rows after cleaning: {rows_after}")
print(f"Rows removed: {rows_before - rows_after}")
```

I performed two main data cleaning steps. First, I checked for and removed duplicate rows using the drop_duplicates() function. Second, I checked for and removed rows with missing values (NaN) using the dropna() function.

The cleaninng was needed because I found missing values,
and missing values were a problem because empty cells cannot be used in calculations like averages and would cause errors when creating charts and visualizations.

changes made:
Before cleaning, the dataset had 391 rows. After removing rows with missing values, 182 rows remain. This means 209 rows were removed because they had incomplete information. The cleaned dataset now has no missing values, making it ready for accurate analysis.

##Univariate Analysis

###Distribution of Restaurant Ratings

```python
# Extract numeric ratings from the 'rate' column
df_clean['rate_clean'] = df_clean['rate'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)
plt.figure(figsize=(8,5))
sns.histplot(df_clean['rate_clean'], bins=20)
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Restaurants")
plt.show()
```

Why histogram: Shows how ratings are spread across all restaurants

Main insight: Most restaurants have ratings between 3.6-4.2 stars

Pattern: Few restaurants have very low ratings (below 3.5)

What we learn: Most restaurants in Bengaluru have good quality

###Distribution of Cost for Two People

```python
df_clean['cost_clean'] = df_clean['approx_cost(for two people)'].astype(str).str.extract(r'(\d+)').astype(float)
plt.figure(figsize=(8,5))
sns.histplot(df_clean['cost_clean'], bins=30)
plt.title("Distribution of Cost for Two People")
plt.xlabel("Cost (Rupees)")
plt.ylabel("Number of Restaurants")
plt.show()
```

Why histogram: Shows the price range of restaurants

Main insight: Most restaurants cost 300-800 rupees for two people

Pattern: Few very expensive restaurants (above 800)

What we learn: Bengaluru has variety dining options

## Bivariate Analysis

### Online Delivery vs Number of Restaurants

```python
plt.figure(figsize=(6,5))
online_counts = df_clean['online_order'].value_counts()
online_counts.plot(kind='bar')
plt.title("Online Delivery vs Number of Restaurants")
plt.xlabel("Online Order Available")
plt.ylabel("Number of Restaurants")
plt.show()
```

Why bar chart: Compares two categories clearly

Main insight: Most restaurants offer online delivery

Pattern: "Yes" bar is much taller than "No" bar

What we learn: Food delivery apps are very popular in Bengaluru

###Table Booking vs Number of Restaurants



```python
plt.figure(figsize=(6,5))
booking_counts = df_clean['book_table'].value_counts()
booking_counts.plot(kind='bar')
plt.title("Table Booking vs Number of Restaurants")
plt.xlabel("Table Booking Available")
plt.ylabel("Number of Restaurants")
plt.show()
```

Why bar chart: Compares restaurants with/without table booking

Main insight: Most restaurants DO NOT offer table booking surprisingly

Pattern: "No" bar is much taller than "Yes" bar

What we learn: Walk-ins are accepted at most places, reservations not needed

###Cost vs Rating Relationship

```python
plt.figure(figsize=(8,5))
plt.scatter(df_clean['cost_clean'], df_clean['rate_clean'], alpha=0.5)
plt.title("Cost vs Rating Relationship")
plt.xlabel("Cost for Two People (Rupees)")
plt.ylabel("Rating")
plt.show()
```

Why scatterplot: Shows if expensive restaurants have higher ratings

Main insight: Weak relationship between cost and rating, that means you can find good food at any price range

Pattern: Points are scattered, no clear line

What we learn: Paying more does NOT guarantee better food



##Most Popular Restaurant Types

```python
plt.figure(figsize=(12,6))
rest_type_counts = df_clean['rest_type'].value_counts().head(10)
rest_type_counts.plot(kind='bar', color='green')
plt.title("Most Popular Restaurant Types")
plt.xlabel("Restaurant Type")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Why bar chart: Shows which restaurant types are most common

Main insight: [Casual Dining/Cafe/Quick Bites] is most common

Pattern: People prefer casual, affordable dining

What we learn: Fine dining is less common, for special occasions only

## Correlation Matrix

```python
# Correlation Matrix
numeric_cols = ['rate_clean', 'cost_clean', 'votes']
numeric_df = df_clean[numeric_cols].dropna()

plt.figure(figsize=(6,4))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix (Rating, Cost, Votes)")
plt.show()
```

Why heatmap: Shows relationships between rating, cost, and votes

Main insight: Rating and votes have strongest correlation

Pattern: Cost has weak correlation with rating

What we learn: Good restaurants get more reviews, but expensive doesn't mean better

## LOCATION-BASED ANALYSIS

###Number of Restaurants by Location

```python
plt.figure(figsize=(12,6))
location_counts = df_clean['location'].value_counts().head(10)
location_counts.plot(kind='bar')
plt.title("Number of Restaurants by Location")
plt.xlabel("Location")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Why bar chart: Shows which areas have the most restaurants

Main insight: banashankari has the most restaurants

Pattern: Restaurants are concentrated in certain areas

What we learn: Go to top locations for more dining choices

### Average Rating by Location

```python
plt.figure(figsize=(12,6))
location_rating = df_clean.groupby('location')['rate_clean'].mean().sort_values(ascending=False).head(10)
location_rating.plot(kind='bar', color='orange')
plt.title("Average Rating by Location")
plt.xlabel("Location")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Why bar chart: Shows which locations have highest-rated restaurants

Main insight: Jayanagar has the highest average rating

Pattern: Quality and quantity don't always go together

What we learn: Best food might be in smaller, less crowded areas
